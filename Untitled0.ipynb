```python
from google import genai
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
from google import genai



# Base instructions
base_prompt = """
I want to create a long-form guided sleep meditation divided into 4 parts, each with at least 1200 words. The tone should be slow, gentle, relaxing — similar to Tara Brach or Michael Sealey.

Each part builds upon the last:
Part 1: Grounding and Breath Awareness
Part 2: Letting Go of Tension and Thought
Part 3: Emotional Safety and Inner Compassion
Part 4: Deep Sleep Descent and Subconscious Soothing

Use poetic, spacious language. Guide the listener into rest using soft metaphors and calming affirmations. Avoid intellectual or stimulating language. Write like you are speaking it slowly.
"""

# Part-specific instructions
part_descriptions = {
    1: "Now write **Part 1: Grounding and Breath Awareness**, with at least 1200 words.",
    2: "Now write **Part 2: Letting Go of Tension and Thought**, with at least 1200 words.",
    3: "Now write **Part 3: Emotional Safety and Inner Compassion**, with at least 1200 words.",
    4: "Now write **Part 4: Deep Sleep Descent and Subconscious Soothing**, with at least 1200 words."
}

results = []

# Loop through all parts and generate content
for part_num, part_prompt in part_descriptions.items():
    full_prompt = base_prompt + "\n\n" + part_prompt

    print(f"--- Generating Part {part_num} ---")

    response = client.models.generate_content(
        model="gemini-2.5-flash",  # or use "gemini-1.5-pro" for better quality
        contents=full_prompt
    )

    #print(f"\n===== PART {part_num} =====\n")
    #print(response.text)
    #print("\n===========================\n")
    results.append(response.text)


```

```python




from google import genai
from google.genai import types
import wave

# Set up the wave file to save the output:
def wave_file(filename, pcm, channels=1, rate=24000, sample_width=2):
   with wave.open(filename, "wb") as wf:
      wf.setnchannels(channels)
      wf.setsampwidth(sample_width)
      wf.setframerate(rate)
      wf.writeframes(pcm)



response = client.models.generate_content(
   model="gemini-2.5-flash-preview-tts",
   contents=f"""Say this slowly, gently, and with a soothing meditation voice. Speak calmly, in a relaxed tone, as if guiding someone into deep sleep. Use long pauses between ideas and keep your pace slow and nurturing: {results[1]}""",
   config=types.GenerateContentConfig(
      response_modalities=["AUDIO"],
      speech_config=types.SpeechConfig(
         voice_config=types.VoiceConfig(
            prebuilt_voice_config=types.PrebuiltVoiceConfig(
               voice_name='Kore',
            )
         )
      ),
   )
)

data = response.candidates[0].content.parts[0].inline_data.data

file_name='out1.wav'
wave_file(file_name, data) # Saves the file to current directory
```

```python
from moviepy.editor import AudioFileClip, CompositeAudioClip
from moviepy.audio.AudioClip import concatenate_audioclips

# Load clips
voice_clip = AudioFileClip("/content/out1.wav")
bg_clip = AudioFileClip("/content/background.mp3").volumex(0.2)  # lower background volume

# Adjust background music to match voice duration
if bg_clip.duration < voice_clip.duration:
    # Loop the background to fill the voice duration
    loops = int(voice_clip.duration / bg_clip.duration) + 1
    bg_clip = concatenate_audioclips([bg_clip] * loops)

# Trim background to match voice duration
bg_clip = bg_clip.subclip(0, voice_clip.duration)

# Combine both
final = CompositeAudioClip([bg_clip, voice_clip])
final.write_audiofile("final_output.mp3", codec="libmp3lame", fps=44100)

```
